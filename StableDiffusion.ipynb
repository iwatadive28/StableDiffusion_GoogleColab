## 各種ファイルのダウンロード
数分かかる。

````python
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", use_auth_token=YOUR_TOKEN)
````

````python
pipe.to("cuda")
````

## Stable Diffusionで画像を生成
以下のコマンドを実行すればOK。「指示」の部分には生成して欲しい画像の説明を英語で入力し、「ファイル名」の部分には生成する画像を保存する際の名前を入力します。

```
prompt = "指示"
image = pipe(prompt)["sample"][0]
image.save(f"ファイル名.png")
```

````python
prompt = "Logo about a Technorogy Challenge Team by Eichiro Oda in One Piece style , no background"
fname  = "Logo_about_a_Technorogy_Challenge_Team_by_Eichiro_Oda_in_OnePieceStyle_noBG2.png"
image = pipe(prompt)["sample"][0]
image.save(fname)

from IPython.display import Image
display(image)
````
